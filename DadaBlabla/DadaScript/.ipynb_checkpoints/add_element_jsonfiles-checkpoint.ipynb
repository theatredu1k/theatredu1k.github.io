```python
import json
import random
```

```python
#initiate the json files with some data

data = {'description': 'a list ot seasons for the story', 'seasons': ['winter']}

with open("../dataset/NarrativeSystem/seasons.json", 'w') as json_file:
  json.dump(data, json_file)
```

```python
#user input new element for the json file
new_season = input("Enter another season:")
#add new season to data ['seasons']
data['seasons'].append(new_season)
#check the json files
with open("../dataset/NarrativeSystem/seasons.json", 'w') as json_file:
  json.dump(data, json_file)
print(data)
```
