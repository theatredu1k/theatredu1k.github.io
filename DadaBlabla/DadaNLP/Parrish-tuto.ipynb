```python
import random
import math
```

```python
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

```python
html_src = """
<body>
<p style="height: 240px; background-image: url(https://i.giphy.com/media/OmK8lulOMQ9XO/giphy-downsized.gif);">Hello!</h1>
</body>
</html>
"""
show_html(html_src)
```

```python
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ min-height: 32em; overflow: hidden; }}
    </style>
</head>
<body>
{content}
</body>
</html>"""
```

```python
interp_src = html_tmpl.format(title="My first HTML page", content="<h1>This is me trying Allison Parrish tutorial on concrete composition -html!</h1>")
print(interp_src)
```

```python
show_html(interp_src)
```

```python
open("test.html", "w").write(interp_src)
```

```python
my_div = "<div style='position: absolute; left: 33%; top: 20%;'>hello!</div>"
html_src = html_tmpl.format(title="hello", content=my_div)
show_html(html_src)
```

```python
divs = []
for i in range(1000):
    x = random.random() * 100
    y = random.random() * 100
    this_div = f"<div style='position: absolute; left: {x}%; top: {y}%;'>★ @ o</div>"
    divs.append(this_div)
html_src = html_tmpl.format(title="Stars", content="".join(divs))
```

```python
show_html(html_src)
```

```python
open("test.html", "w").write(html_src)
!open test.html
```

assign a random size and transparency to the stars, using the font-size CSS style, the color CSS style, and the rgba CSS color unit:

```python
divs = []
for i in range(1000):
    x = random.random() * 100
    y = random.random() * 100
    size = random.randrange(4, 32)
    alpha = random.random()
    this_div = f"<div style='position: absolute; left: {x}%; top: {y}%; font-size: {size}pt; color: rgba(0,0,0,{alpha})'>★</div>"
    divs.append(this_div)
html_src = html_tmpl.format(title="Stars of different sizes", content="".join(divs))
show_html(html_src)
```
